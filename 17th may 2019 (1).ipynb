### Update/modify a row in CSV file


```python
def updateRowCSV(filename,rowindex,rowdata):        ## these r database queries-create,add new data,update,delete,retrieve
    df=readCSV(filename)
    df.loc[rowindex]=rowdata
    df.to_csv(filename,index=False)                  # index=false need to use otherwise takes more time for execution
    return df
rowdata=['7','7','04000US08','Florida',123,2334,345,4567,5678,6782,7892,8456,9820]
updateRowCSV('DataFiles/Income.csv',7,rowdata)
#df=df.drop('Unnamed:0',axis=1)
#readCSV('DataFiles/Income.csv')
```

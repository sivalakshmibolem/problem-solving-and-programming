# plotting 2D Graphs using matplotlib

```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
```

```python
plt.plot([1,2,3,4],[1,8,27,64])
plt.xlabel('Time')
plt.ylabel('Distance')
plt.show()
```

### numpy

```python
import numpy as np

s=np.arange(0.0,2.0,0.1)

v = 1+np.sin(2 * np.pi * s)

plt.plot(s,v)
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.grid(True)
plt.show()


```

### plot a line graph on income vs year for the state of california

```python
import pandas as pd
df = pd.read_csv('DataFiles/Income.csv')
y = df.values[4]
x = df.columns
print(y)
print(x)
```

```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

years=[]
for i in range(2,len(x)):
    years.append(x[i])
years
income=[]
for i in range(2,len(y)):
    income.append(y[i])
income
plt.title
plt.plot(years,income)
plt.xlabel('year')
plt.ylabel('income')
plt.grid(True)
plt.show()
```

```python
width=1/2
plt.title("Income of California")
plt.bar(years, income, width, color= 'Gray')
plt.savefig()
```
